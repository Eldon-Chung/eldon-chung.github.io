```python
import numpy as np 
from matplotlib import pyplot as plt 


x = np.arange(1,64) 
y1 = np.log2(x)
y2 = x
plt.xlabel("size of array") 
plt.ylabel("number of comparisons") 
plt.plot(x,y1, 'r-', label="binary search")
plt.plot(x,y2, 'b-',label="linear search") 
plt.legend()
plt.savefig('logvslinear.png')
plt.show()
```

```python
import numpy as np 
from matplotlib import pyplot as plt 


x = np.arange(2,32) 
y1 = 9 * np.log2(x)
y2 = 2 * x + 3
plt.xlabel("value of b") 
plt.ylabel("number of operations") 
plt.plot(x,y1, 'r-', label="algorithm 2")
plt.plot(x,y2, 'b-',label="algorithm 1") 
plt.legend()
plt.savefig('fast-exp.png')
plt.show()
```
