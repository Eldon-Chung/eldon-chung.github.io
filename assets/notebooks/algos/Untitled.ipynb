```python
import numpy as np 
from matplotlib import pyplot as plt 


x = np.arange(1,1000) 
y1 = x ** 2
y2 =  x ** 1.585
plt.xlabel("value of b") 
plt.ylabel("number of operations") 
plt.plot(x,y1, 'r-', label="naive algorithm")
plt.plot(x,y2, 'b-', label="Karatsuba's algorithm") 
plt.legend()
plt.savefig('karatsuba.png')
plt.show()
```
